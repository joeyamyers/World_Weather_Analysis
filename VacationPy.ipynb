```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
```

```python
# Store the CSV you saved created in part one into a DataFrame.
city_data_df = pd.read_csv("weather_data/cities.csv")
city_data_df
```

```python
# data we use for mapping must be an int or float
city_data_df.dtypes
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Heatmap of temperature
# Get the latitude and longitude.
locations = city_data_df[["Lat", "Lng"]]
# Get the maximum temperature.
max_temp = city_data_df["Max Temp"]
# Assign the figure variable.
fig = gmaps.figure()
# adjust heatmap zoom, intensity, & point radius
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
# Assign the heatmap variable.
# use max() and list comprehension to cancel out any negative max temps
heat_layer = gmaps.heatmap_layer(locations, weights=[max(temp, 0) for temp in max_temp], dissipating=False, max_intensity=300, point_radius=4)
# Add the heatmap layer.
fig.add_layer(heat_layer)
# Call the figure to plot the data.
fig
```

```python
# Get Vacation criteria: a feature that allows customers to search for locs based on temp prefs
# Ask the customer to add a minimum and maximum temperature value.
min_temp = float(input("What is the minimum temperature you would like for your trip?"))
max_temp = float(input("What is the maximum temperature you would like for your trip?"))
```

```python
preferred_cities_df = city_data_df.loc[(city_data_df["Max Temp"] >= min_temp) & (city_data_df["Max Temp"] <= max_temp)]
preferred_cities_df
```

```python
# once they have filtered the database based on temp pref, show them a heatmap for max temp for filtered cities
# create a marker for ea city that will display name of city/country code/max temp & name of neaby hotel
# within 3 miles of coordinates when clicked
```

```python
# need a new DF for hotels bc customers will need it to filter
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = preferred_cities_df[["City", "Country", "Max Temp", "Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df
```

```python
# now using lat/lng & google places API retrieve hotels from a nearby search

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
hotel_params = {
    "type": "lodging",
    "radius": 5000,
    "key": g_key}

# use iterrows to iterate through pandas dataframe
for index, row in hotel_df.iterrows():
    
     # get latitude/longitude from df
    lat = row['Lat']
    lng = row['Lng']

    # add location to params dict
    hotel_params['location'] = f"{lat},{lng}"
    
    # make request and get JSON data from search
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    hotels = requests.get(base_url, params=hotel_params).json()
    
    # Use a try/except to get the first hotel from the results and store the name for each row
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except:
        print("Hotel not found... skipping.")
```
